```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("outputs(Qwen3-0.6B)/results.csv")

metrics = ["tfidf_cosine", "rouge1_f1", "rouge2_f1", "rougeL_f1"]
corr_matrix = df[metrics].corr()

plt.figure(figsize=(6, 5))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="Blues",   # blue-only version of your example
    vmin=0.0,       # do NOT start from -1
    vmax=1.0,
    linewidths=0.5,
    cbar=True
)

plt.title("Correlation heatmap: TF-IDF & ROUGE (Qwen3-0.6B)")
plt.tight_layout()
plt.savefig(
    "outputs(Qwen3-0.6B)/metric_correlation_heatmap1.png",
    dpi=300
)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load results
df = pd.read_csv("outputs(Qwen3-0.6B)/results.csv")

x = df["sample_idx"].values
nli_scores = df["hallucination_nli"].values
heuristic_scores = df["hallucination_heuristic"].values

# Same structure as original code
fig2, ax2 = plt.subplots(figsize=(6, 4))

# Marker-only plots (no connecting lines)
ax2.plot(
    x,
    nli_scores,
    marker="o",
    linestyle="None",
    label="NLI hallucination"
)

ax2.plot(
    x,
    heuristic_scores,
    marker="s",
    linestyle="None",
    label="Heuristic hallucination"
)

ax2.set_xlabel("Sample index")
ax2.set_ylabel("Hallucination score")
ax2.set_title("Two hallucination metrics across samples")
ax2.set_ylim(0.0, 1.0)

ax2.grid(True)
ax2.legend()

fig2.tight_layout()
fig2.savefig(
    "outputs(Qwen3-0.6B)/hallucination_points_only.png",
    dpi=300
)

plt.close(fig2)

print("Saved hallucination point-only plot")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Paths to results
path_qwen = "outputs(Qwen3-0.6B)_GPT/results.csv"
path_reranker = "outputs(Qwen3-Reranker-8B)_GPT/results.csv"

# Load data
df_qwen = pd.read_csv(path_qwen)
df_reranker = pd.read_csv(path_reranker)

# X axis
x_qwen = df_qwen["sample_idx"].values
x_reranker = df_reranker["sample_idx"].values

# GPT hallucination scores
gpt_qwen = df_qwen["hallucination_gpt4"].values
gpt_reranker = df_reranker["hallucination_gpt4"].values

# Create single figure
fig, ax = plt.subplots(figsize=(8, 4))

# Qwen3-0.6B (blue points)
ax.plot(
    x_qwen,
    gpt_qwen,
    marker="o",
    linestyle="None",
    color="tab:blue",
    label="Qwen3-0.6B"
)

# Qwen-Reranker-8B (orange points)
ax.plot(
    x_reranker,
    gpt_reranker,
    marker="o",
    linestyle="None",
    color="tab:orange",
    label="Qwen-Reranker-8B"
)

ax.set_title("GPT hallucination across samples")
ax.set_xlabel("Sample index")
ax.set_ylabel("Hallucination score")
ax.set_ylim(0.0, 1.0)

ax.legend()
ax.grid(True, alpha=0.3)

fig.tight_layout()
fig.savefig(
    "outputs/gpt_hallucination_points_both_models.png",
    dpi=200
)
plt.close(fig)

print("Saved GPT hallucination comparison (both models, points only)")
```
